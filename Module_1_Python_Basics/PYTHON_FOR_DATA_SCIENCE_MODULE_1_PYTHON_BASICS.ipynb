```python
# STRING OPERATIONS
# In Python, a string is a sequence of characters. A string is contained within two quotes:
"MICHAEL JACKSON"
# You could also use single quotes. A string can be spaces, or digits
"1 2 3 4 5 6 "
# A string can also be special characters.
" # @ % & * "
# We can bind or assign a string to another
NAME = "MICHAEL JACKSON"
# t is helpful to think of a string as an ordered sequence.
# Each element in the sequence can be accessed using an index represented by the array of numbers.
# To access the first index we have
NAME[0]
```

```python
# We can also use negative indexing with strings.
# The last element is given by the index -1. The first element can be obtained by index
NAME[-1]
```

```python
# -1. The first element can be obtained by index -15, and so on
NAME[-15]
```

```python
# STRING: SLICING
# String slicing means taking (extracting) a part of a string using its index.
NAME[0:4]
```

```python
# STRING: STRIDE
# Stride (also called step) means how many characters to skip when slicing a string.
# We can also input a stride value as follows.
NAME[::2]
# The 2 indicates we select every second variable.
```

```python
# The General Syntax for stride: string[start:end:stride]
# start = where to begin (included)
# end = where to stop (not included)
# stride = how many characters to move forward each time
# In this example after the first letter we skip to pick the third letter and so on

NAME = "MICHAEL JACKSON"
NAME[::3]
```

```python
# STRING: SLICING & STRIDE TOGETHER
# We can also incorporate slicing into stride.
# In this case.
# start = where to begin (included) = 0
# end = where to stop (not included) = 5
# stride = how many characters to move forward each time = 2
NAME[0:5:2]
```

```python
# STRING: LEN
# len() is a built-in Python function that returns the number of items (length) in a string, list, tuple, or other collections.
len("MICHAEL JACKSON")
print(len(NAME))
```

```python
# STRING CONCATENATION
# String concatenation means joining (combining) two or more strings together.
# We can concatenate or combine strings. We use the addition symbols.
STATEMENT = NAME + "  IS THE BEST "
STATEMENT
```

```python
# STRING MULTIPLICATION
# We can replicate values of a string. We simply multiply the string by the number
# of times we would like to replicate it, in this case, three.
3 * "MICHAEL JACKSON"
```

```python
# STRINGS ARE IMMUTABLE
# This means you cannot change the value of the string, but you can create a new string. For example, you can create a new string by
# setting it to the original variable and concatenated with a new string.
NAME = "MICHAEL JACKSON"
NAME = NAME + " IS THE BEST"
NAME
```

```python
# STRINGS: ESCAPSE SEQUENCE
# An escape sequence is a combination of characters that starts with a backslash \ and represents a special character in a string.
# They are used when you cannot type certain characters directly or when they have a special meaning in Python.
# \\ : Backslash
# \' : Single quote
# \" : Double quote
# \n : New line
# \t : Tab (space)
# \b : Backspace
# \r : Carriage return
# \f : Form feed
```

```python
print("michael jackson \n is the best")
```

```python
print("michael jackson \t is the best")
```

```python
# STRING METHOD
# String methods are built-in functions used to work with text (strings).
# When we apply a method to the string "A”,we get a new string “B" that is different from "A".
```

```python
# METHOD:UPPER
# This method converts lower case characters to upper case characters.
A = "thriller is the sixth studio album"
B = A.upper()
B
```

```python
# METHOD: REPLACE
# The method replaces a segment of the string, i.e., a substring with a new string.
# We input the part of the string we would like to change.
# The second argument is what we would like to exchange the segment with.
# The result is a new string with the segment changed.
A = "Michael Jackson is the best"
B = A.replace('Michael', 'Janet')
B
```

```python
# METHOD: FIND
# The method find, finds sub-strings. The argument is the sub-string you would like to find. The output is the first index of the sequence.
# GENERAL SYNTAX:string.find(substring, start, end)
# String :The text you want to search inside
# Substring:The letter or word you are looking for
# Start: The position where python should start searching
# End: The position where python should stop searching
A = "Michael Jackson is the best"
B = A.find('el')
B
```

```python
A.find("best",2,5)
# What does this mean?
# A.find("best", 2, 5) means:
# Look for the substring "best"
# Start searching at index 2
#Stop searching before index 5
# Between index 2 and 5, the characters are:c h a
# The word "best" is NOT in this range — it appears much later in the sentence.The output is -1
# -1 means: “not found in this range.”
```

```python
A.find("best",2,28)
```
